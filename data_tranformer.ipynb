```python
# Imports üì•

# Packages
#-----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Settings
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # to see all columns in DataFrames


# Data 
# -----------------------------------------------------------------------
abc_coorp_df = pd.read_csv("HR RAW DATA.csv", index_col=0)
```

```python
# DataTransformer class definition ‚úçÔ∏è

class DataTransformer:
    def __init__(self, dataframe):
        self.df = dataframe

    def replace_gender_values(self):
        """Replaces the values in the Gender column with 'Male' and 'Female'."""
        self.df['Gender'] = self.df['Gender'].replace({0: 'Male', 1: 'Female'})

# the following methods were suggeted by chatGPT. REVIEW, EDIT OR DELETE

    def convert_to_numeric(self, column_name):
        """Converts a column to numeric type."""
        self.df[column_name] = pd.to_numeric(self.df[column_name], errors='coerce')

    def remove_duplicates(self):
        """Removes duplicate rows."""
        self.df.drop_duplicates(inplace=True)

    def fix_negative_distances(self):
        """Corrects negative values in the DistanceFromHome column."""
        self.df['DistanceFromHome'] = self.df['DistanceFromHome'].abs()

    def correct_typographical_errors(self):
        """Corrects typographical errors in the MaritalStatus column."""
        self.df['MaritalStatus'] = self.df['MaritalStatus'].replace({'Marreid': 'Married'})

    def drop_redundant_columns(self):
        """Drops redundant columns like 'SameAsMonthlyIncome'."""
        self.df.drop(columns=['SameAsMonthlyIncome'], inplace=True, errors='ignore')

    def get_dataframe(self):
        """Returns the transformed DataFrame."""
        return self.df


```
